# Combinación de registros de presencia (DwC) con fichas de especies (PLIC)

**Bibliotecas de Python**

```python
import pandas as pd
```

**Entradas**

```python
# Archivo con registros de presencia
INPUT_SPECIES_OCCURRENCES = "input/occurrences.csv"

# Archivo con fichas de especies
INPUT_SPECIES_PAGES = "input/plinian_core.txt"

# Archivo con taxonomía de las fichas de especies
INPUT_SPECIES_PAGES_TAXON = "input/taxon.txt"
```

**Salidas**

```python
# Archivo de salida
OUTPUT_SPECIES_OCCURRENCES_PAGES = "output/species_occurrences_pages.csv"
```

**Procesamiento**

```python
# Carga de dataframes
input_species_occurrences_df = pd.read_csv(INPUT_SPECIES_OCCURRENCES, sep='\t', encoding='utf_8')
input_species_pages_df = pd.read_csv(INPUT_SPECIES_PAGES, sep='\t', encoding='utf_8')
input_species_pages_taxon_df = pd.read_csv(INPUT_SPECIES_PAGES_TAXON, sep='\t', encoding='utf_8')
```

```python
# Filtrado de la taxonomía de las páginas de especies
species_pages_taxon_df = input_species_pages_taxon_df[input_species_pages_taxon_df['class'] == 'Aves']

species_pages_taxon_df
```

```python
# Merge de fichas de especies y taxonomía
species_pages_df = pd.merge(species_pages_taxon_df[['taxonID','class','order','family','genus', 'scientificName']], 
                            input_species_pages_df[['id','HabitatUnstructured']], 
                            left_on='taxonID', 
                            right_on='id',
                            how='inner')

species_pages_df
```

```python
species_occurrences_df = input_species_occurrences_df

# Filtrado de registros de presencia (opcional)
species_occurrences_df = input_species_occurrences_df[input_species_occurrences_df['class'] == 'Aves']
# species_occurrences_df = species_occurrences_df[species_occurrences_df['basisOfRecord'] == 'PRESERVED_SPECIMEN']

species_occurrences_df
```

```python
# Dataframe de salida
species_occurrences_pages_df = pd.merge(species_occurrences_df,
                                        species_pages_df[['scientificName','HabitatUnstructured']], 
                                        left_on='species', 
                                        right_on='scientificName',
                                        how='left')

species_occurrences_pages_df
```

```python
# Exportación a CSV
species_occurrences_pages_df.to_csv(OUTPUT_SPECIES_OCCURRENCES_PAGES, index = False)
```
